**Dataset Pre-processing**

```python
#Cleaning null value
arr = dataset.select_dtypes(include = (['int64', 'float64'])).columns
#print(arr)
for c in arr:
    dataset[c].fillna(dataset[c].median(), inplace=True)
#dataset
arr.isna()
col = dataset.select_dtypes(include=('object')).columns
#print(col)
for c in col:
    dataset[c].fillna(dataset[c].mode().values[0], inplace=True)

dataset.select_dtypes(include=('object')).isna().sum()
```

```python
dataset
```

```python
dataset["Sex"].replace("M",0,inplace = True)
dataset["Sex"].replace("F",1,inplace = True)
dataset["Stage"].replace(4,"x",inplace = True)
dataset["Stage"].replace(1,0,inplace = True)
dataset["Stage"].replace(2,0,inplace = True)
dataset["Stage"].replace(3,0,inplace = True)
dataset["Stage"].replace("x",1,inplace = True)
dataset["Hepatomegaly"].replace("Y",1,inplace = True)
dataset["Hepatomegaly"].replace("N",0,inplace = True)
dataset["Ascites"].replace("Y",1,inplace = True)
dataset["Ascites"].replace("N",0,inplace = True)
dataset["Edema"].replace("Y",1,inplace = True)
dataset["Edema"].replace("N",0,inplace = True)
dataset["Edema"].replace("S",0,inplace = True)
dataset["Status"].replace("CL",0,inplace=True)
dataset["Status"].replace("C",1,inplace=True)
dataset["Status"].replace("D",0,inplace=True)
dataset["Drug"].replace("D-penicillamine",1,inplace = True)
dataset["Drug"].replace("Placebo",0,inplace = True)
dataset["Spiders"].replace("Y",1,inplace = True)
dataset["Spiders"].replace("N",0,inplace = True)

dataset
```

```python
from sklearn.preprocessing import MinMaxScaler


list_of_features = ["N_Days", "Age", "Bilirubin", "Cholesterol", "Albumin", "Copper","Alk_Phos","SGOT","Tryglicerides","Platelets","Prothrombin"]

scaler = MinMaxScaler()

dataset[list_of_features] = scaler.fit_transform(dataset[list_of_features])
```

```python
dataset.head()
```

```python
x_data = dataset[['ID', 'N_Days', 'Status', 'Drug',
       'Age', 'Sex', 'Ascites', 'Hepatomegaly',
       'Spiders', 'Edema','Bilirubin','Cholesterol','Albumin','Copper','Alk_Phos','SGOT','Tryglicerides','Platelets','Prothrombin']]
y_data = dataset[["Stage"]]
```

```python
x_data
```

```python
y_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_data, y_data,
                                                    test_size = 0.2,
                                                    random_state = 1,
                                                    stratify = y_data)
print("Training data",X_test)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
dataset.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
```

```python
X_train.shape
```

```python
from sklearn.metrics import accuracy_score

lr=accuracy_score(y_test, y_pred)
print(lr)
```
